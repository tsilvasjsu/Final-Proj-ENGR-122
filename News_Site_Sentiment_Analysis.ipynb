###Article 1: Order to reopen Alcatraz - CNN coverage.

##Split paragraphs into sentences

```r
# Here is the syntax to split paragraphs into sentences. Can be used as a reference for the remainder of the articles.

# Split paragraphs into sentences

# Article 1 - CNN
cnn_sentences1 <- get_sentences(cnn_paragraphs1)
# Print with a header
cat("CNN ARTICLE 1 SENTIMENT ANALYSIS \n")
cat("------------------------------------\n")
head(cnn_sentences1)

# Article 1 - Democracy Now
dn_sentences1 <- get_sentences(dn_paragraphs1)
# Print with a header
cat("DEMOCRACY NOW ARTICLE 1 SENTIMENT ANALYSIS \n")
cat("------------------------------------------\n")
head(dn_sentences1)

# Article 1 - Fox
fox_sentences1 <- get_sentences(fox_paragraphs1)
# Print with a header
cat("FOX ARTICLE 1 SENTIMENT ANALYSIS \n")
cat("------------------------------------\n")
head(fox_sentences1)
```

##Get Sentiments for Articles

* **get_sentiment**: calculates sentiment scores with values ranging from negative to positive (basic sentiment analysis).
* **get_nrc_sentiment**: uses NRC lexicon which classifies text into different emotion categories (e.g., anger, fear, joy).

```r
# Get_sentiment for all articles:

# Article 1 - CNN coverage
cnn_sentiment1 <- get_sentiment(cnn_sentences1)
cat("----------------------------------------------\n")
cat("CNN ARTICLE 1 SENTIMENT ANALYSIS \n")
cat("----------------------------------------------\n")
summary(cnn_sentiment1)

# Article 1 - Democracy Now coverage
dn_sentiment1 <- get_sentiment(dn_sentences1)
cat("----------------------------------------------\n")
cat("DEMOCRACY NOW ARTICLE 1 SENTIMENT ANALYSIS \n")
cat("----------------------------------------------\n")
summary(dn_sentiment1)

# Article 1 - Fox coverage
fox_sentiment1 <- get_sentiment(fox_sentences1)
cat("----------------------------------------------\n")
cat("FOX ARTICLE 1 SENTIMENT ANALYSIS \n")
cat("----------------------------------------------\n")
summary(fox_sentiment1)
```

```r
# Get_NRC_sentiments for all articles:

# Article 1 - CNN coverage
cnn_sentiment1.nrc <- get_nrc_sentiment(cnn_sentences1)
# Print with a title
cat("CNN ARTICLE 1 NRC SENTIMENT ANALYSIS\n")
cat("---------------------------------------\n")
head(cnn_sentiment1.nrc)

# Article 1 - Democracy Now coverage
dn_sentiment1.nrc <- get_nrc_sentiment(dn_sentences1)
# Print with a title
cat("DEMOCRACY NOW ARTICLE 1 NRC SENTIMENT ANALYSIS\n")
cat("-----------------------------------------------\n")
head(dn_sentiment1.nrc)

# Article 1 - Fox coverage
fox_sentiment1.nrc <- get_nrc_sentiment(fox_sentences1)
# Print with a title
cat("FOX ARTICLE 1 NRC SENTIMENT ANALYSIS\n")
cat("-----------------------------------------------\n")
head(fox_sentiment1.nrc)
```
